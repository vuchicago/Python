```
###########CORRELATION BETWEEN VARIABLES AND PRICE
from scipy.stats.stats import pearsonr
from matplotlib import pyplot as plt
```

```
a=print("Volume = ", pearsonr(df[vol].astype(float),df[dep_var].astype(float)))
corr=pd.DataFrame()
pd.concat([corr,a])
for i in ind_var:
  a=print(i," = ",pearsonr(df[i].astype(float),df[dep_var].astype(float)))
  a=pd.DataFrame(a)
corr=pd.concat([corr,a])

```

```
df_new=df.sample(frac=1,random_state=3) ###Samples data with same random state
df_train=df_new[:np.int(.8*len(df_new))]
df_val=df_new[np.int(.8*len(df_new)):]
```

```
df_train.describe()
```

```
############################################THE EFFECTS OF VOLUME ON PRICE GIVEN EACH VARIABLE IN THE DEP_VAR LIST training dataset
corr=pd.DataFrame()

for i in ind_var:
  df_train_var=df_train[df_train[i]==1]
  a=print("Effects of Vol on Price given Vol:",df_train_var[vol].sum()," and ",i," = ",pearsonr(df_train_var[vol].astype(float),df_train_var[dep_var].astype(float)))
  a=pd.DataFrame(a)
corr=pd.concat([corr,a])  
```

```
############################################THE EFFECTS OF VOLUME ON PRICE GIVEN EACH VARIABLE IN THE DEP_VAR LIST validation dataset
corr=pd.DataFrame()

for i in ind_var:
  df_val_var=df_val[df_val[i]==1]
  a=print("Effects of Vol on Price given Vol:",df_val_var[vol].sum()," and ",i," = ",pearsonr(df_val_var[vol].astype(float),df_val_var[dep_var].astype(float)))
  a=pd.DataFrame(a)
corr=pd.concat([corr,a])  
```
